```python
#### Importing all required libraries
import re
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

#### Installing chrome Driver based on current version 
#### And Create a new instance of google chrome
browser = webdriver.Chrome(ChromeDriverManager().install())

mobiles_list = []
for i in range(1, 42):
    #### Access google chrome and open flipkart website
    browser.get('https://www.flipkart.com/search?q=phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(i))
    
    #### Extracting the data using class name
    result=browser.find_elements_by_xpath((".//div[@class='E2-pcE _1q8tSL']"))
    result1=result[1].find_elements_by_xpath((".//div[@class='_2pi5LC col-12-12']"))
    for mobile in result1[:-2]:
        mobile_dict = {}
        #### Extracting 'Name' of the mobile
        mobile_dict['Name'] = mobile.find_element_by_class_name('_4rR01T').text
        
        #### Extracting 'Rating' of the mobile
        try:
            mobile_dict['Rating'] = mobile.find_element_by_class_name('_3LWZlK').text
        except:
            pass
        
        #### Extracting 'Discount Price' of the mobile
        try:
            mobile_dict['Discount Price'] = mobile.find_element_by_xpath((".//div[@class='_30jeq3 _1_WHN1']")).text
        except:
            pass
        
        #### Extracting 'Actual Price' of the mobile
        try:
            mobile_dict['Actual Price'] = mobile.find_element_by_xpath((".//div[@class='_3I9_wc _27UcVY']")).text
        except:
            pass
        
        #### Extracting 'Discount percentage' of the mobile
        try:
            mobile_dict['Discount %'] = mobile.find_element_by_class_name('_3Ay6Sb').text
        except:
            pass
        
        #### Extracting Features of the mobile
        try:
            feature = str(mobile.find_element_by_class_name('_1xgFaf').get_attribute('innerHTML'))
            mobile_dict['Features'] = re.sub("<[^>]+>", "\n", feature).split('\n')[1::2]
        except:
            pass
        
        mobiles_list.append(mobile_dict)
        
#### Closing chrome browser
browser.close()

#### Converting all the extracted elements into dataframe
mobiles = pd.DataFrame(mobiles_list)

#### Printing top 5 rows of the data
mobiles.head()
```

```python
#### Saving the data in to csv file
mobiles.to_csv('mobiles.csv')
```
