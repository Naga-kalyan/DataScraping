```python
#### Importing all required libraries
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

#### Installing chrome Driver based on current version 
#### And Create a new instance of google chrome
browser = webdriver.Chrome(ChromeDriverManager().install())

list_grotal = []
for j in range(1,6):
    #### Access google chrome and open grotal website
    browser.get('https://www.grotal.com/Kurukshetra/music-C55A0P'+str(j)+'A0/')
    #### Extracting the data using class name
    result = browser.find_elements_by_class_name('result-row')
    for i in range(0,len(result)):
        if((i != 1) and (i != 3)):
            dict_grotal = {}
            
            #### Extracting 'name' from grotal website
            try:
                dict_grotal['name'] = result[i].find_elements_by_class_name('cursor')[0].text
            except:
                dict_grotal['name'] = None
            
            #### Extracting 'area' from grotal website
            try:
                dict_grotal['area'] = result[i].find_elements_by_class_name('cursor')[1].text
            except:
                dict_grotal['area'] = None
                
            #### Extracting 'address' from grotal website
            try:
                dict_grotal['address'] = result[i].find_elements_by_css_selector('div')[5].text
            except:
                dict_grotal['address'] = None
            
            #### Extracting 'number' from grotal website
            try:
                dict_grotal['number'] = result[i].find_elements_by_class_name('ph-no')[0].text.replace(u'\xa0',u' ')
            except:
                dict_grotal['number'] = None
            
            #### Extracting 'services' from grotal website
            try:
                dict_grotal['services'] = result[i].find_elements_by_css_selector('div')[8].text
            except:
                dict_grotal['services'] = None
            list_grotal.append(dict_grotal)

#### Closing chrome browser
browser.close() 

#### Converting all the extracted elements into dataframe
grotal = pd.DataFrame(list_grotal)

#### Printing top 5 rows of the data
grotal.head()
```

```python
#### Saving the data in to csv file
grotal.to_csv('grotal.csv')
```
