```python
#### Importing all required libraries
import time
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

#### Installing chrome Driver based on current version 
#### And Create a new instance of google chrome
browser = webdriver.Chrome(ChromeDriverManager().install())
browser.maximize_window()
base_url="https://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s="
houses_list = []
for page in range(0,30,10):
    #### Requesting the server to get the content of the web page
    browser.get(base_url+str(page)+".html")
    time.sleep(5)
    #### Extracting table data using class name
    houses = browser.find_elements_by_class_name("propertyRow")
    for house in houses:
        houses_dict = {}
        
        #### Extracting 'Address' of the house
        houses_dict['Address'] = house.find_elements_by_class_name('propAddressCollapse')[0].text
        
        #### Extracting 'Locality' of the house
        houses_dict["Locality"] = house.find_elements_by_class_name('propAddressCollapse')[1].text
        
        #### Extracting 'Price' of the house
        houses_dict["Price"] = house.find_element_by_class_name('propPrice').text.replace("\n","")
        
        #### Extracting info of the 'Beds' in a house
        try:
            houses_dict["Beds"] = house.find_element_by_class_name('infoBed').find("b").text
        except:
            pass
        
        #### Extracting 'Area' of the house
        try:
            houses_dict["Area"] = house.find_element_by_class_name('infoSqFt').find("b").text
        except:
            pass
        
        #### Extracting info of the 'Full Baths' in a house
        try:
            houses_dict["Full Baths"] = house.find_element_by_class_name('infoValueFullBath').find("b").text
        except:
            pass
        
        #### Extracting 'info of the 'Half Baths' in a house
        try:
            houses_dict["Half Baths"] = house.find_element_by_class_name('infoValueHalfBath').find("b").text
        except:
            pass
        
        #### Extracting 'Property Description' of the house
        try:
            houses_dict["PROPERTY DESCRIPTION"] = house.find_element_by_class_name('propertyDescription').text[23:-1]
        except:
            pass
        
        #### Extracting 'Features' of the house
        for columnGroup in house.find_elements_by_class_name('columnGroup'):
            for feature_group in columnGroup.find_elements_by_class_name('featureGroup'):
                
                #### Extracting 'Lot Size' of the house
                if "Lot Size" in feature_group.text:
                    houses_dict["Lot Size"] = columnGroup.find_element_by_class_name('featureName').text
                
                #### Extracting 'Age' of the house
                if "Age" in feature_group.text:
                    houses_dict["Age"] = columnGroup.find_element_by_class_name('featureName').text
                
                #### Extracting 'Appliances' of the house
                if "Appliances:" in feature_group.text:
                    feature_Appliances = ''
                    for featureName in columnGroup.find_elements_by_class_name('featureName'):
                        feature_Appliances += featureName.text
                        houses_dict["Appliances"] = feature_Appliances
                    
                #### Extracting 'Basement' of the house
                if "Basement" in feature_group.text:
                    houses_dict["Basement"] = columnGroup.find_element_by_class_name('featureName').text
        houses_list.append(houses_dict)

#### Closing chrome browser
browser.close()   

#### Converting all the extracted elements into dataframe
houses_df = pd.DataFrame(houses_list) 

#### Printing top 5 rows of the data
houses_df.head(5) 
```

```python
#### Saving the data in to csv file
houses_df.to_csv('houses.csv')
```
