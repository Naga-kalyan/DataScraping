```python
#### Importing all required libraries
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

#### set the maximum column width to unlimited
pd.set_option('display.max_colwidth', None)

#### Installing chrome Driver based on current version 
#### And Create a new instance of google chrome
browser = webdriver.Chrome(ChromeDriverManager().install())

list_quotes = []
for page_no in range(1,6):
    #### Access google chrome and open quotes website
    browser.get('http://quotes.toscrape.com/page/'+str(page_no)+'/')
    
    #### Extracting all quotes using class name
    quotes = browser.find_elements_by_class_name('quote')
    for i in range(len(quotes)):
        dict_quote = {}
        #### Extracting 'About' of the quote
        dict_quote["About"] = 'http://quotes.toscrape.com'+quotes[i].find_elements_by_tag_name("a")[0].get_attribute("href")
        
        #### Extracting 'Author' of the quote
        dict_quote["Author"] = quotes[i].find_element_by_class_name('author').text
        
        #### Extracting 'Quote'
        dict_quote["Quote"] = quotes[i].find_element_by_class_name('text').text.replace('“',"").replace('”',"")
        
        #### Extracting 'Tags' of the quote
        dict_quote["Tags"] = ', '.join(quotes[i].find_element_by_class_name('tags').text.split()[1:])
        list_quotes.append(dict_quote)

#### Closing chrome browser
browser.close()

#### Converting all the extracted elements into dataframe
quotes = pd.DataFrame(list_quotes)

#### Printing top 5 rows of the data
quotes.head()
```

```python
#### Saving the data in to csv file
quotes.to_csv('quotes.csv')
```
