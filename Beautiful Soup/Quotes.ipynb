```python
#### Importing all required libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
list_quotes = []
for page_no in range(1,6):
    #### Requesting the server to get the content of the web page
    page = requests.get('http://quotes.toscrape.com/page/'+str(page_no)+'/')
    
    #### Parsing the html
    soup = BeautifulSoup(page.content,'html.parser')
    
    #### Extracting all quotes using class name
    quotes = soup.find_all(class_='quote')
    for i in range(len(quotes)):
        dict_quote = {}
        #### Extracting 'About' of the quote
        dict_quote["About"] = 'http://quotes.toscrape.com'+quotes[i].find_all("a")[0].attrs['href']
        
        #### Extracting 'Author' of the quote
        dict_quote["Author"] = quotes[i].find(class_='author').text
        
        #### Extracting 'Quote'
        dict_quote["Quote"] = quotes[i].find(class_='text').text.replace('“',"").replace('”',"")
        
        #### Extracting 'Tags' of the quote
        dict_quote["Tags"] = ';'.join(quotes[i].find(class_='tags').text.split('\n')[3:-1])
        list_quotes.append(dict_quote)

#### Converting all the extracted elements into dataframe
quotes = pd.DataFrame(list_quotes)

#### Printing top 5 rows of the data
quotes.head()
```

```python
#### Saving the data in to csv file
quotes.to_csv('quotes.csv')
```
