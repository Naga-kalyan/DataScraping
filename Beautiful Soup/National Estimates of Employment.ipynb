```python
#### Importing all required libraries
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup

#### Requesting the server to get the content of the web page
page = requests.get('https://www.bls.gov/web/empsit/cesbmart.htm')

#### Parsing the html
soup = BeautifulSoup(page.content,'html.parser')

#### Extracting table data using class name
table  = soup.find(class_='regular')

#### Extracting table row elements into list
table_body = table.find('tbody')
table_tr = table_body.find_all('tr')

list_seasonally = []
for month_seasonal in table_tr:
    dict_seasonally={}
    dict_seasonally['Month'] = month_seasonal.text.split('\n')[1].strip()
    dict_seasonally['Levels-As Revised'] = month_seasonal.text.split('\n')[2]
    dict_seasonally['Levels-As Previously Published'] = month_seasonal.text.split('\n')[3]
    dict_seasonally['Levels-Difference'] = month_seasonal.text.split('\n')[4]
    dict_seasonally['Over-the-month Changes-As Revised'] = month_seasonal.text.split('\n')[5]
    dict_seasonally['Over-the-month Changes-As Previously Published'] = month_seasonal.text.split('\n')[6]
    dict_seasonally['Over-the-month Changes-Difference'] = month_seasonal.text.split('\n')[7]
    list_seasonally.append(dict_seasonally)

#### Converting all the extracted elements into dataframe
data_seasonally = pd.DataFrame(list_seasonally) 

#### Printing the data
data_seasonally   
```

```python
#### Saving the data in to csv file
data_seasonally.to_csv('seasonally.csv')
```
