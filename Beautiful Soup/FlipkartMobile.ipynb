```python
#### Importing all required libraries
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup

mobiles_list = []
for i in range(1,42):
    #### Requesting the server to get the content of the web page
    page=requests.get('https://www.flipkart.com/search?q=phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(i))
    
    #### Parsing the html
    soup=BeautifulSoup(page.content,'html.parser')
    
    #### Extracting the data using class name
    result=soup.find_all(class_='E2-pcE _1q8tSL')
    result1=result[1].find_all(class_='_2pi5LC col-12-12')
    for mobile in result1[:-2]:
        mobile_dict = {}
        #### Extracting 'Name' of the mobile
        mobile_dict['Name'] = mobile.find(class_='_4rR01T').text
        
        #### Extracting 'Rating' of the mobile
        try:
            mobile_dict['Rating'] = mobile.find(class_='_3LWZlK').text
        except:
            pass
        
        #### Extracting 'Discount Price' of the mobile
        try:
            mobile_dict['Discount Price'] = mobile.find(class_='_30jeq3 _1_WHN1').text
        except:
            pass
        
        #### Extracting 'Actual Price' of the mobile
        try:
            mobile_dict['Actual Price'] = mobile.find(class_='_3I9_wc _27UcVY').text
        except:
            pass
        
        #### Extracting 'Discount percentage' of the mobile
        try:
            mobile_dict['Discount %'] = mobile.find(class_='_3Ay6Sb').text
        except:
            pass
        
        #### Extracting Features of the mobile
        try:
            feature = str(mobile.find(class_='_1xgFaf'))
            mobile_dict['Features'] = re.sub("<[^>]+>", "\n", feature).split('\n')[2:-2:2]
        except:
            pass
        
        mobiles_list.append(mobile_dict)

#### Converting all the extracted elements into dataframe
mobiles = pd.DataFrame(mobiles_list)

#### Printing top 5 rows of the data
mobiles.head()
```

```python
#### Saving the data in to csv file
mobiles.to_csv('mobiles.csv')
```
