```python
#### Importing all required libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
list_grotal = []
for j in range(1,6):
    page = requests.get('https://www.grotal.com/Kurukshetra/music-C55A0P'+str(j)+'A0/')
    soup = BeautifulSoup(page.content,'html.parser')
    result = soup.find_all(class_='result-row')
    for i in range(0,len(result)):
        if((i != 1) and (i != 3)):
            dict_grotal = {}
            try:
                dict_grotal['name'] = result[i].find_all(class_='cursor')[0].text
            except:
                dict_grotal['name'] = None
            try:
                dict_grotal['area'] = result[i].find_all(class_='cursor')[1].text
            except:
                dict_grotal['area'] = None
            try:
                dict_grotal['address'] = result[i].find_all(class_='address gl-listing-new-list')[0].text
            except:
                dict_grotal['address'] = None
            try:
                dict_grotal['number'] = result[i].find_all(class_='ph-no')[0].text.replace(u'\xa0',u' ')
            except:
                dict_grotal['number'] = None
            try:
                dict_grotal['services'] = result[i].find_all(class_='desc gl-listing-new-list')[0].text
            except:
                dict_grotal['services'] = None
            list_grotal.append(dict_grotal)

#### Converting all the extracted elements into dataframe
grotal = pd.DataFrame(list_grotal)

#### Printing top 5 rows of the data
grotal.head()
```

```python
#### Saving the data in to csv file
grotal.to_csv('grotal.csv')
```
